```python
import pandas as pd
```

```python
pd.__file__
```

```python
# DATA DOWNLOAD
prefix = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/"
url = f"{prefix}/yellow_tripdata_2021-01.csv.gz"
url
```

```python
# DATA TYPES AND COLUMNS FOR PARSING
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]
```

```python
# ACTUAL DOWNLOAD AND READ INTO DF
# df = pd.read_csv(url, nrows = 100)
df = pd.read_csv(url, 
                 dtype=dtype, 
                 parse_dates=parse_dates)
df.head()
```

```python
len(df)
```

```python
# ENGINE FOR DB CONNECTION
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
# SHOWING SCHEMA USED BELOW IN .to_sql() METHOD
print(pd.io.sql.get_schema(df, name="yellow_taxi_data", con=engine))
```

```python
df.head(0).to_sql(name="yellow_taxi_data", con=engine, if_exists="replace")
```

```python
# GETTING DATA IN CHUNKS FOR CHUNK LOADING --> ITERATOR
df_iter = pd.read_csv(url, 
                 dtype=dtype, 
                 parse_dates=parse_dates,
                iterator=True,
                chunksize=100000)
# GETS ME AN INTERATOR, NOT DF
```

```python
from tqdm.auto import tqdm
for chunk in tqdm(df_iter):
    print(len(chunk))
    chunk.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")
    print("Chunk inserted")
```
